```python
import pandas as pd
from arcgis.features import GeoAccessor

df = pd.read_parquet("../data/raw/paris_accidents_raw.parquet")

# Tell ArcGIS which column is the geometry
sdf = GeoAccessor.from_df(df, geometry_column="SHAPE")

sdf.head()
```

```python
sdf.columns
```

```python
cols_to_keep = [
    "id_accident",            # unique accident ID
    "annee",                  # year
    "age",                    # victim age
    "sexe_victime",           # victim sex
    "victime_type",           # pedestrian / driver / passenger...
    "categorie",              # user / vehicle category
    "gravite",                # TARGET: severity of outcome
    "milieu",                 # environment (urban, tunnel, etc.)
    "nom_arrondissement",     # Paris district
    "longitude", "latitude",  # coordinates for spatial work
    "blesses_legers",
    "blesses_hospitalises",
    "tues",                   # fatalities indicator
    "iris",                   # neighborhood code
    "SHAPE"                   # geometry for ArcGIS
]
```

```python
sdf = sdf[cols_to_keep].copy()
```

```python
sdf.head()
```

```python
sdf.to_parquet("../data/processed/paris_accidents_raw.parquet")
```
