```python
import pandas as pd
from arcgis.features import GeoAccessor

df = pd.read_parquet("../data/processed/paris_accidents_cleaned.parquet")

# Tell ArcGIS which column is the geometry
sdf = GeoAccessor.from_df(df, geometry_column="SHAPE")

sdf.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
sdf["age"] = scaler.fit_transform(sdf[["age"]])
```

```python
sdf.head()
```

```python
gravite_map = {
    "Blessé léger": 0,
    "Blessé hospitalisé": 1,
    "Tué": 2
}
sdf['gravite'] = sdf['gravite'].map(gravite_map)
```

```python
sdf.head()
```

```python
sdf.dtypes
```

```python
import pandas as pd

# 1) convert to datetime
sdf['annee'] = pd.to_datetime(sdf['annee'], errors='coerce')

# 2) keep only the year as integer
sdf['annee'] = sdf['annee'].dt.year.astype('Int64')   # nullable int
# or, if you are sure there are no NaT left:
# sdf['annee'] = sdf['annee'].dt.year.astype('int64')

```

```python
sdf['longitude'] = pd.to_numeric(sdf['longitude'], errors='coerce')
sdf['latitude']  = pd.to_numeric(sdf['latitude'],  errors='coerce')
```
